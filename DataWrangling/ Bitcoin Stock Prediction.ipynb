## Calculate what the highest and lowest opening prices were for the stock in this period.

```python
p = dict['dataset']['data']
z= [x[1] for x in p]
res=[]
for val in z:
    if val!= None:
        res.append(val)
print("The maximum opening value in 2018 & 2020 was " +  str(max(res)))
print("The minimum opening value in 2018 & 2020 was " +  str(min(res)))
```
